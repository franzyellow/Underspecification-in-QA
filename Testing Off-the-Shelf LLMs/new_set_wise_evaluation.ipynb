```python
def source_set_evaluation(input_df, source_set_list):
  for source_set in source_set_list:
    df_source = input_df[input_df['direct_source'] == source_set]
    print(f"Printing {source_set} classification report in results:")
    print(classification_report(df_source["gold_judgment"], df_source['model_pred']))

def source_set_evaluation_path(results_path, source_set_list):
  path_heuristic = results_path + "/*.csv"
  for file in glob.glob(path_heuristic):
    df_all = pd.read_csv(file)
    for source_set in source_set_list:
      df_source = df_all[df_all['direct_source'] == source_set]
      print(f"Printing {source_set} classification report in {file}:")
      if 'direct_label_transform' in df_source.columns:
        print(classification_report(df_source["gold_judgment"], df_source['direct_label_transform'], labels=["fully specified", "underspecified"],
    target_names=["fully specified", "underspecified"]))
      else:
        print(classification_report(df_source["gold_judgment"], df_source['model_pred']))


source_set_list = ['AmbigQA', 'IN3', 'CoCoNot', 'CLAMBER']
```

```python
ambig_qa=['AmbigQA']
source_set_evaluation_path(exp_1_best, ambig_qa)
```

```python
source_set_evaluation_path(exp_1_best, source_set_list)
```
